```python
import boto3
import botocore
from pprint import pprint

session = boto3.session.Session(profile_name="default")
```

```python
def get_accoun_id():
    """This function prints account ID && acount arn
    Parameter: None
    Return None
    """
    try:
        client = session.client('sts')
        response = client.get_caller_identity()
        print(f"My acount id is: {response.get('Account')}\nAccount arn: {response.get('Arn')}")
    except Exception as err:
        print(err)
    return None 

get_accoun_id()
```

## Using boto3 to work with iam

```python
resource = session.resource('iam') # This is global
client = session.client('iam')

def list_roleswithClient():
    """This function list all available roles with client. 
    Parameter: None
    Return: None
    """
    try:
        response = client.list_roles()['Roles']
        for role in response:
            print(f"Role arn: {role.get('Arn')}\nRole Name: {role.get('RoleName')}\nRole Description: {role.get('Description')}")
    except Exception as err:
        print(err)   
    return None  

list_roleswithClient()
```

### Catching Access denied err using botocore exception 

```python
def list_bucket():
    try:
        session = boto3.session.Session(profile_name="default")
        resource = session.resource(service_name='s3')
        buckets = resource.buckets.all()
        for bucket in buckets:
            print(bucket.name)
    except botocore.exceptions.ClientError as err:
        if err.response.get('Error')['Code'] == 'AccessDenied':
            print("You don't access to list buckets")
        else:
            print(err.response.get('Error')['Code'])
    except Exception as e:
        print(e)
    return None  
```

```python
list_bucket()
```
